```python
# ['从西方吹来的风' '从东南方吹来的风' '从东方吹来的风' '从西北方吹来的风' '从东南偏东方向吹来的风' '从南方吹来的风'
#  '从西北偏西方向吹来的风' '从西南偏西方向吹来的风' '从东南偏南方向吹来的风' '从西北偏北方向吹来的风' '从东北方吹来的风'
#  '从北方吹来的风' '从西南方吹来的风' '从西南偏南方向吹来的风' '从东北偏东方向吹来的风' '从东北偏北方向吹来的风' '无风' nan]
def DD_to_N(dd):
    direction={
        '北':'0/',
        '东':'4/',
        '南':'8/',
        '西':'12/'
    }
    if isinstance(dd, str):
        for char, num in direction.items():
            dd = dd.replace(char, num)
        numbers = re.findall('\d+', dd)
        numbers = [int(num) for num in numbers]
        if len(numbers)==0:
            result=-1
        else:
            result=numbers[0]
            for i in range(len(numbers) - 1):
                result = (result + numbers[i + 1]) / 2
    else:
        result=dd
    return result

merged_df['DDN'] = merged_df['DD'].apply(DD_to_N)
merged_df
```
